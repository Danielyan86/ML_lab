```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values (range)
x = np.linspace(-10, 10, 400)

# Define the sigmoid functions with different parameters
y1 = 1 / (1 + np.exp(-5 * x))
y2 = 1 / (1 + np.exp(-0.1 * x))
y3 = 1 / (1 + np.exp(-0.3 * x))

# Plot the functions
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='1 / (1 + exp(-5 * x))', color='red')
plt.plot(x, y2, label='1 / (1 + exp(-0.1 * x))', color='blue')
plt.plot(x, y3, label='1 / (1 + exp(-0.3 * x))', color='green')

# Add title and labels
plt.title('Sigmoid Functions with Different Parameters', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)

# Add legend
plt.legend()

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values (range)
x = np.linspace(-2, 2, 400)

# Define the e^x function
y = np.exp(x)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='e^x', color='blue')

# Add title and labels
plt.title('Plot of the Exponential Function $e^x$', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('$e^x$', fontsize=12)

# Add legend
plt.legend()

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```
